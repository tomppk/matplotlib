## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))
              
# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("../sample-plot.png")
```

```python
# 0. Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [10, 23, 50, 60, 79]
y = [100, 200, 300, 600, 800]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Cloud thickness and rain amount",
      xlabel="Cloud thickness",
      ylabel="Rain amount")

# 5. Save & show plot
fig.savefig("../cloud-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# Create some data
x =np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x, y)
ax.set(title="Dan's Nut Butter Store",
     ylabel="Price ($)");
```

```python
# Create data
x = np.arange(0, 9999, 100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Plot from Python dictionary
fruit_inventory = {"Watermelon": 342,
                  "Pineapple": 455,
                  "Coconut": 179}

# Create plot
fig, ax = plt.subplots()
ax.bar(fruit_inventory.keys(), fruit_inventory.values())
ax.set(title="Fruit Depot inventory",
      ylabel="Quantity");
```

```python
fig, ax = plt.subplots()
ax.barh(list(fruit_inventory.keys()), list(fruit_inventory.values()))
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));

ax1.set(title="Degrees",
       ylabel="Balance factor")
ax2.set(title="Snowball toss",
       ylabel="Accuracy")
ax3.set(title="Nut butter store",
       ylabel="Prices ($)")
ax4.set(title="Obesity per 100k",
       ylabel="Kilograms");
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));

```
